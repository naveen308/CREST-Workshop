```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
def extract_prompt(example):
    # Assume the prompt ends at the first newline (typical format)
    chosen_lines = example["chosen"].split("\n")
    rejected_lines = example["rejected"].split("\n")

    prompt_line = chosen_lines[0] if chosen_lines[0] == rejected_lines[0] else ""
    return {
        "prompt": prompt_line,
        "chosen": example["chosen"],
        "rejected": example["rejected"]
    }

dataset_with_prompt = dataset.map(extract_prompt)
```

```python
safety_keywords = ["bomb", "chemical", "phishing", "weapon", "explosive"]

def contains_safety_keywords(example):
    text = example["chosen"].lower() + " " + example["rejected"].lower()
    return any(keyword in text for keyword in safety_keywords)

# Load a streaming version of the dataset to iterate quickly
#from datasets import load_dataset
from itertools import islice

#dataset = load_dataset("Anthropic/hh-rlhf", split="train", streaming=True)

# Filter safety-related samples
safety_samples = filter(contains_safety_keywords, dataset_with_prompt)

# Take first 5 safety samples for inspection
safety_data = list(islice(safety_samples, 10))

# Show results
for i, item in enumerate(safety_data):
    print(f"\nExample {i+1}")
    print("Chosen:", item["chosen"])
    print("Rejected:", item["rejected"])
```

```python
print(f"Total examples after filtering: {len(safety_data)}")
```

```python
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
```

```python
model.config.pad_token_id = tokenizer.pad_token_id
```

```python
model = GPT2LMHeadModel.from_pretrained(model_name)
model = model.to(device)
```

```python
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model.config.pad_token_id = tokenizer.pad_token_id
```

```python
dpo_data = Dataset.from_list(safety_data)
```

```python
dpo_config = DPOConfig(
    output_dir="./dpo_safety_result",
    per_device_train_batch_size=2,
    learning_rate=5e-7,
    num_train_epochs=3,
    gradient_accumulation_steps=4,
    logging_steps=10,
    save_steps=500,
    remove_unused_columns=False,
    beta=0.1,
    loss_type="sigmoid",
    fp16=False,
    bf16=False  # Added to explicitly disable bf16
)
```

```python
trainer = DPOTrainer(
    model=model,
    args=dpo_config,  # Pass DPOConfig as args
    train_dataset=dpo_data,
    tokenizer=tokenizer,
)

print("DPO Trainer setup complete.")
```

```python
trainer.train()
```

```python
def generate_response(prompt_text):
    inputs = tokenizer(prompt_text, return_tensors="pt")
    inputs = {k: v.to(device) for k, v in inputs.items()}  # Move inputs to device
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=20,
            num_return_sequences=1, # Sampling Strategies: Beam Search, Nucleus Sampling....
            temperature=0.7,
            pad_token_id=tokenizer.pad_token_id
        )
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

```

```python
# Example usage (before training, this will give baseline responses)
print("\nExample generation (before training):")
prompt = "Can i make bomb at home?"
response = generate_response(prompt)
print(f"Prompt: {prompt}")
print(f"Response: {response}")
```
